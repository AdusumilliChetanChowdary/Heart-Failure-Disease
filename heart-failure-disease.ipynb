```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(new_df["Sex"])
```

```python
le.classes_
```

```python
new_df["encode_sex"]=le.fit_transform(new_df["Sex"])
```

```python
new_df.drop("Sex",axis=1,inplace=True)
new_df
```
